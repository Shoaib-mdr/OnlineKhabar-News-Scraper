```python
#CONVERSION OF Request Object Into Beautiful SOup Obj
soup = bs4.BeautifulSoup(requestURL.text, 'lxml')
```

```python
#LIST ALL CATEGORIES
CategoriesTag = soup.find_all("li", class_=re.compile("menu-item menu-item-type-post_type menu-item-object-page menu-item|menu-item menu-item-type-taxonomy menu-item-object-category menu-item"), id=False)
CategoriesTag
```

```python
categoryList = []
categoryLink = []
for eachCategoryTag in CategoriesTag:
    categoryList.append(eachCategoryTag.find("a", href=True).text)
    headLink = eachCategoryTag.find("a", href=True)
    categoryLink.append(headLink['href'])
```

```python
#CONVERSION INTO SOUP OBJECT
def convertToSoup(req):
    soup = bs4.BeautifulSoup(req.text, 'lxml')
    return soup
```

```python
#SCRAPPER
def scrapper(soup):
    newsTitleList = []
    newsLinkList = []
    head_links = soup.find_all("h2", class_="title__small post__title")
    for eachHeadLinks in head_links:
        for a in eachHeadLinks.find_all("a", href=True):
            newsLinkList.append(a['href'])
            newsTitleList.append(a.text)
    print(newsLinkList, newsTitleList)
    return newsLinkList, newsTitleList
```

```python
#MAIN ALGO
for i in range(0,len(categoryLink)):
    
    reqCategoryURL = requests.get(categoryLink[i])
    reqCategoryURL.encoding = 'utf-8'
    #Convesion to soup obj
    soup = convertToSoup(reqCategoryURL)
    newsLinkList, newsTitleList = scrapper(soup)
    print(categoryList[i])
    os.mkdir('C:\\Users\\Dell\\Documents\\Projects\\WebScraping\\Data\\'+categoryList[i])
    for j in range(0, len(newsTitleList)):
        stringText = []
        
        reqNews = requests.get(newsLinkList[j])
        reqNews.encoding = 'utf-8'
        soupNews = convertToSoup(reqNews)
        headText = soupNews.find_all("div", class_="col colspan3 main__read--content ok18-single-post-content-wrap")
        for eachHeadText in headText:
            text = eachHeadText.find_all("p")
            stringText = ""
            for k in text:
                stringText+=k.text
        print(stringText)
        filename=str(j)+"shoaib.txt"
        print(filename)
        f= open("C:\\Users\\Dell\\Documents\\Projects\\WebScraping\\Data\\"+categoryList[i]+"\\"+ filename,"w+",encoding="utf-8")
        f.write("TITLE: " + "\t" + newsTitleList[j] + "\n" + "NEWS: " + "\t" + stringText)
        f.close()
  
      
    
    
```
